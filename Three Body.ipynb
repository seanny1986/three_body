## 2. Background
### 2.1 The Three-Body Problem
For the uninitiated, the three-body problem dates back to the time of Sir Isaac Newton, and deals with the dynamics of three bodies acting under the gravitational influence of one another. Whereas the two-body problem is well studied and has known analytic solutions, the three-body problem is highly chaotic, with relatively few stable orbital systems. To date, several families (in fact, 16 [2]) of stable solutions have been discovered using numerical methods, and in 1912, Karl Sundman published a general solution to the problem [1] that takes the form of a convergent series. Unfortunately, Sundman's solution -- while an incredible achievement -- is of little practical value, as it relies on an impossibly large number of terms to be useful.

So what is this problem? From [Wikipedia](https://en.wikipedia.org/wiki/Three-body_problem), our model of the three-body system takes the form:

$\frac{dV_1}{dt} = -Gm_2\frac{r_1 - r_2}{|r_1 - r_2|^3}-Gm_3\frac{r_1 - r_3}{|r_1 - r_3|^3}$

$\frac{dV_2}{dt} = -Gm_1\frac{r_2 - r_1}{|r_2 - r_1|^3}-Gm_3\frac{r_2 - r_3}{|r_2 - r_3|^3}$

$\frac{dV_3}{dt} = -Gm_2\frac{r_3 - r_2}{|r_3 - r_2|^3}-Gm_1\frac{r_3 - r_1}{|r_3 - r_1|^3}$

Where $G$ is the gravitational constant, and $m_n$ denotes the mass of the $n_{th}$ body. The presence of the third body makes this problem far more complex than the case for only two bodies. Our hammer for modelling and making predictions about systems with three or more bodies is simulation using numerical integrators -- that is, for each object, we calculate the force acting on it, find its acceleration, and then update its velocity and position using a small time increment. Mathematically:

$V_{t+dt} \approx V_{t} + \frac{dV_{t}}{dt}\Delta t$

$X_{t+dt} \approx X_{t} + \frac{V_{t+dt}}{dt}\Delta t$

Where $\Delta t$ is a small, finite timestep (as opposed to the infinitesimal $dt$). We incur some error with an update over a finitely small time interval, but by making our time step small, this error will be small enough that we can unroll our system over long time horizons and maintain a reasonable degree of accuracy. In general, numerical simulation of orbital systems doesn't conserve energy (due to the aforementioned error); traditionally, this is dealt with by using so-called symplectic integrators, though in more recent works (e.g. [2]), it's common to use adaptive stepsize methods. Since Scipy doesn't include any symplectic integrators (and since the fixed stepsize would smash my poor old macbook) I'll use an adaptive method and lean on the modern miracle that is Scipy.

Finally, I'll make a quick simplification to our system of equations so that I'm not trying to wrangle my way through a second PhD thesis. I'll assume that our three bodies all have the same mass, which -- given that our gravitational constant is (exactly as its name would suggest) constant -- lets us look at the simpler system:

$\frac{dV_1}{dt} = -k\left(\frac{r_1 - r_2}{|r_1 - r_2|^3}+\frac{r_1 - r_3}{|r_1 - r_3|^3}\right)$

$\frac{dV_2}{dt} = -k\left(\frac{r_2 - r_1}{|r_2 - r_1|^3}+\frac{r_2 - r_3}{|r_2 - r_3|^3}\right)$

$\frac{dV_3}{dt} = -k\left(\frac{r_3 - r_2}{|r_3 - r_2|^3}+\frac{r_3 - r_1}{|r_3 - r_1|^3}\right)$

where I'll refer to $k$ as the stiffness. This is a fairly common simplification to make. This way, we're simulating the time evolution of *some* system without trying to justify what that system is. If we need to recover parameters for a physical system that might exhibit the stable orbits that we find, we can always do that later.

### 2.2 Projections
In order to find our hypothetical 2-simplex, we need to first understand what a projection is. We take this for granted, but we see it every day on our phones and laptops -- a projection is an operation that transforms a 3D vector or set of coordinates into a 2D representation that can be shown on a screen. Typically, this is done using the following operation:

$u_w = \begin{bmatrix} x_w  \\ y_w \\ z_w \\ w \end{bmatrix} = \mathbf{A} \begin{bmatrix} x_s  \\ y_s \\ z_s \\ w_s \end{bmatrix} = \mathbf{A} u_s$

$u_p = \frac{1}{w}\begin{bmatrix} x_w  \\ y_w \\ z_w \end{bmatrix}$

Where $\mathbf{A}$ is a $4 \times 4$ projection matrix that encodes our object's rotation and translation (and sometimes shear) relative to the space we're projecting into (though, depending on application, it can also encode other variables, such as camera calibration). When we do a projection, we flatten one of those dimensions (in our case, we'll call it $z$), which is what lets us project it onto our screen. We lose a bit of information in the process (i.e. reversing the operation is more difficult), but that's usually not important in the types of applications where projection is used. Typically, we add a clipping value $w$ into our system, and we project into *clip space* first. From there, we use this clipping constant to normalize the other coordinates (this is the equivalent of projecting into a coordinate system that is parallel with the camera, and *then* projecting to the camera itself -- it makes the math a bit more straightforward).

What we're primarily concerned with in this problem is the opposite of a projection (the inverse projection), where we take set of 2D coordinates, and find a matrix transform that maps them to a set of 3D coordinates, which -- as we've already mentioned -- is somewhat more difficult. The reason this is difficult is that for a given set of 2D cordinates for our triangle, there are an infinite number of 2-simplexes of varying dimensions and distances that could be a valid solution to the problem. Consider, for example, that we could have two objects $A$ and $B$, where $A$ is twice the size of $B$. If we position $A$ twice as far from the screen as $B$, it will project the same shadow, making it an equally valid solution. Now consider that there are an infinite combination of sizes and distances that we could choose!

Needless to say, this is a problem, and it barely scratches the surface. For example, we also have to consider that different orientations could also cast the same shadow, and shearing transformations could give us wildly different solutions in 3D space that correspond to the same solution in 2D space. Oof.

Well, let's have a go anyway, since there are ways that we can reduce the space a bit. Since we aren't particularly interested in finding $A$ (what we actually want to find is its inverse $A^{-1}$), we can set this problem up as: 

$u_{w^{'}} = \begin{bmatrix} x_{w^{'}}  \\ y_{w^{'}} \\ z_{w^{'}} \\ {w^{'}} \end{bmatrix} = \mathbf{A^{-1}} \begin{bmatrix} x_p  \\ y_p \\ z_p \\ w_p \end{bmatrix} = \mathbf{A^{-1}} u_p$

$u_s = \frac{1}{w^{'}}\begin{bmatrix} x_{w^{'}}  \\ y_{w^{'}} \\ z_{w^{'}} \end{bmatrix}$

Where $w^{'}$ is the inverse clip space (for more info see here). Thankfully, the inverse projection that we find isn't particularly important -- that is, we don't care where in 3D space our object is oriented, because we can always choose a new datum point and translate our object. Rather, what we're interested in is that our inverse projection is *consistent* over time -- that is, for small changes in our three-body system, the inverse projection shouldn't map to wildly different inverse projections -- so that we can observe the behaviour of our simplex over the same time period.

When we do the opposite, and try to solve the inverse problem of rotating and translating our 2-simplex in 3D space, and and projecting it into 2D, we'll need to do a perspective projection from our 3D space onto a 2D plane. Thankfully, we don't need to find $\mathbf{A}$ for that either. Instead, we will use:

$\begin{bmatrix} x_p  \\ y_p \end{bmatrix} = m\begin{bmatrix} x_s  \\ y_s \end{bmatrix}$

$m = c/z_s$

for some chosen constant $c$. This will result in an arbitrary projection onto a 2D plane that we can then visualize. As with our inverse projection, the coordinates themselves shouldn't matter, because again, we're mainly interested in the evolution of the system over time. I'm not convinced that if we have what looks to be a valid solution that the coordinates will make a difference. We should be able to find "some" set of parameters (stiffness, initial velocities) to make our system work within this coordinate frame (though maybe someone with a deeper understanding can help expand on this).

Armed with this knowledge, I'll go through the methodology below.

## 3. Methodology
I'm not a 3D graphics expert, so I don't know what techniques are typically used to convert 2D projected points into 3D coordinates for a known object. What I do know, however, is some machine learning. We know the 2D coordinates of our 3 "planets" in the $xy$-plane, and we know that an inverse projection takes the form given previously. What we really need to do is find a set of constraints that will let us find consistent inverse projections over long periods of time, given that we have a previous transform that we think is pretty accurate. The following are probably a good start:

1. All sides of the triangle have a fixed length $R=1$ (for convenience).
2. The angle between all edges is $\pi/3$.
3. We can pick an arbitrary projection to project our found 3D points back into 2D, which helps constrain the problem because we can map individual vertices to one another in both 2D and 3D.
4. Our solution at each timestep is likely to be similar to the solution for the previous timestep. We can constrain our optimization routine to find answers that are "close" to whatever our previous answer was using the previous surface normal vector.

With this information, I came up with the following methology:

1. Guess $A^{-1}$, either using a random matrix, or $\mathbf{I}$. If we already have a solution from a previous timestep, we use that instead.
2. If we have a previous solution, assume that the normal vector $\hat{n}$ will be close to the new solution.
2. Add small noise $\epsilon$ to $A^{-1}$ to ensure normal vectors aren't collinear.
2. Calculate $u_f$ using the above equations.
3. Find all edge lengths $e$, and all angles $a$.
4. Project $u_f$ back into 2D using our chosen projection function.
4. Find the cost of our inverse projection matrix using $\sum_i (e_i - e_{g})^2 + (a_i - a_g)^2 + (v_{p,i} - v_{p,i}^{*})^2$, where $e_g$ is our target edge length $R$, $a_g$ is our target angle of $\pi/3$, and $v_{p,i}$ and $v_{p,i}^{*}$ are our 2D projected coordinates from our inverse projected 3D coordinates, and the ground truth, respectively.
5. Use an optimizer to find the inverse projection matrix that minimizes this cost function.

It's not pretty, but it's straightforward, and it should hopefully work. We can also consider adding other constraints on the COM of the simplex to help things out a bit. Since our object is rotating and translating in 3D space, we need to repeat this operation at every timestep in order to get the evolution of its position and attitude. This is actually a good thing; since we're taking small timesteps, our previous solution should be a good guess for our new solution, which will speed things up considerably. Furthermore, it should help ensure that we get solutions that are consistent with the previous timestep. Once we have the 3D coordinates of our simplex's vertices, we can find a surface normal vector by taking the cross product over two of the edge vectors, and normalizing it. This way, we can characterize not only the translation of the object (measured using the position of the centre of mass), but also its attitude.

### 4.1 Three-body Simulation
The below code imports all of the modules we'll need, and implements our 2D three-body simulation to be called with scipy.integrate.

```python
import numpy as np
from scipy.optimize import minimize
from scipy.integrate import ode

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.gridspec import GridSpec
from matplotlib import animation

from math import sqrt, acos, asin, pi, sin, cos, radians, degrees

from IPython.display import HTML
from IPython.display import clear_output
from IPython.display import Video

        
def three_body_2d(t, z, k):
    """
    Returns gradients for the 2D three-body problem, to be called with a scipy ODE integrator.
    
    Args:
        t (float) : time variable
        z (ndarray) : 1D array containing the state in a [positions, velocities] format
        k (float) : a stiffness parameter that determines the response of the system
    
    Returns:
        z_dot (ndarray) : 1D array containing the time derivatives of the state
    """
    
    params = int(len(z)/2)
    bodies = int(params/2)
    step = 2
    pos = z[:params]
    vel = z[params:]
    r1 = [pos[int(step*i):int(step*(i+1))] for i in range(bodies)]
    r1_dot = [vel[int(step*i):int(step*(i+1))] for i in range(bodies)]
    vecs = [p - q for p in r1 for q in r1 if not np.array_equal(p, q)]
    mags = [np.linalg.norm(v) for v in vecs]
    a = [-k*v/(m**3) for v, m in zip(vecs, mags)]
    r2_dot = [np.sum(np.vstack(a[int(step*i):int(step*(i+1))]), axis=0) for i in range(bodies)]
    z_dot = np.hstack(r1_dot+r2_dot)
    return z_dot
```

### 4.2 Projection and Optimization Functions

```python
from scipy.optimize import NonlinearConstraint

def get_angle(p, q):
    angle = acos(np.sum(p * q/(np.linalg.norm(p)*np.linalg.norm(q))))
    return angle

def get_com(v1, v2, v3):
    return (v1 + v2 + v3) / 3

def get_simplex2(v1, v2, v3):
    com = get_com(v1, v2, v3)
    k1 = com - v1
    k2 = com - v2
    normal = np.cross(k1, k2)
    normal /= np.linalg.norm(normal)
    return [v1, v2, v3], com, normal

def get_simplex(x1, x2, x3, mat):
    v1 = inverse_projection(x1, mat)
    v2 = inverse_projection(x2, mat)
    v3 = inverse_projection(x3, mat)
    com = get_com(v1, v2, v3)
    k1 = com - v1
    k2 = com - v2
    normal = np.cross(k1, k2)
    normal /= np.linalg.norm(normal)
    return [v1, v2, v3], com, normal

def get_vma(X1_p, X2_p, X3_p, mat):
    v1 = inverse_projection(X1_p, mat)
    v2 = inverse_projection(X2_p, mat)
    v3 = inverse_projection(X3_p, mat)
    vs = [v1, v2, v3]
    edges = [p - q for p in vs for q in vs if not np.array_equal(p, q)]    
    mags = [np.linalg.norm(e) for e in edges]
    a1 = get_angle(edges[0], edges[1])
    a2 = get_angle(edges[2], edges[3])
    a3 = get_angle(edges[4], edges[5])
    angles = [a1, a2, a3]
    return vs, mags, angles
    
def get_inverse_projection_matrix(X1_p, X2_p, X3_p, theta, n=None, c=5):
    """
    Optimizes our inverse projection matrix. If n is None, we look for any projection
    that meets our cost function. If n exists, we bound our new solution to be within
    a few degrees of n, as this is the most likely solution.
    
    Args:
        X1_p (ndarray) : a size (2,) array containing the xy location of vertex 1
        X2_p (ndarray) : a size (2,) array containing the xy location of vertex 2
        X3_p (ndarray) : a size (2,) array containing the xy location of vertex 3
        theta (ndarray) : a size (4,4) inverse projection matrix
        n (ndarray) : a size (3,) guess of the surface normal unit vector
        c (float) : a bound in degrees on the deviation of the new surface normal
    
    Returns:
        ans.x (ndarray) : found theta values, reshaped to size (4,4) array
    """
    
    def cost(theta, X1_p, X2_p, X3_p):
        mat = theta.reshape(4,4)
        _, com, _ = get_simplex(X1_p, X2_p, X3_p, mat)
        vs, mags, angles = get_vma(X1_p, X2_p, X3_p, mat)
        xs = [X1_p, X2_p, X3_p]
        cost_1 = sum([(m - 1)**2 for m in mags])
        cost_2 = sum([(a - pi/3)**2 for a in angles])
        cost_3 = sum(np.sum((projection(v) - x)**2) for v, x in zip(vs, xs))
        return cost_1 + cost_2 + cost_3
    
    def cons_f(theta):
        mat = theta.reshape(4,4)
        _, _, norm = get_simplex(X1_p, X2_p, X3_p, mat)
        try:
            return -(get_angle(norm, n) - c * pi/180)
        except:
            return -(0 - c * pi/180)
    
    if n is not None:
        nlc = {'type': 'ineq', 'fun': cons_f}
        ans = minimize(cost, theta.flatten(), args=(X1_p, X2_p, X3_p), constraints=[nlc], tol=1e-10)
    else:
        ans = minimize(cost, theta.flatten(), args=(X1_p, X2_p, X3_p), tol=1e-10)
    return ans.x.reshape(4,4)



def projection(p, f=2.5):
    """
    Projects 3D point onto a 2D plane using a simplified perspective projection.
    
    Args:
        p (ndarray) : a size (3,) array representing a point in 3D space
        f (optional, float) : a scaling constant for the distance to the projection surface.
    
    Returns :
        xy (ndarray) : a size (2,) array representing a point in 2D space.
    """
    _, _, z = p
    k = f/z
    return k * p[:2]

def inverse_projection(p, mat):
    """
    "Unprojects" a 2D point to a point in a 3D space using an inverse projection matrix.
    
    Args:
        p (ndarray) : a size (2,) array representing a point in 2D space
        mat (ndarray) : a size (4,4) array representing an inverse projection matrix
    
    Returns:
        xyz (ndarray) : a size (3,) array representing a point in 3D space
    """
    p = np.append(p, np.ones((2,)))
    q = mat.dot(p.reshape(-1,1)).flatten()
    return q[0:-1]/q[-1]

def rotate_around_axis(u, theta):
    """
    Performs rotation of theta radians around a normalized axis vector u.
    
    Args:
        u (ndarray) : a size (3,) unit vector representing axis of rotation
        theta (float) : a float value in radians giving the rotation angle theta
    
    Returns:
        R (ndarray) : a size (3,3) rotation matrix 
    """
    return np.array([[cos(theta) + u[0]**2 * (1-cos(theta)), 
             u[0] * u[1] * (1-cos(theta)) - u[2] * sin(theta), 
             u[0] * u[2] * (1 - cos(theta)) + u[1] * sin(theta)],
            [u[0] * u[1] * (1-cos(theta)) + u[2] * sin(theta),
             cos(theta) + u[1]**2 * (1-cos(theta)),
             u[1] * u[2] * (1 - cos(theta)) - u[0] * sin(theta)],
            [u[0] * u[2] * (1-cos(theta)) - u[1] * sin(theta),
             u[1] * u[2] * (1-cos(theta)) + u[0] * sin(theta),
             cos(theta) + u[2]**2 * (1-cos(theta))]])

def get_aor(V1s, V2s):
    """
    Calculate axis of rotation and angle of rotation theta using the points in
    V1, V2. This was horrible to get working.
    
    Args:
        V1s (list) : a length N list containing initial points from our object (higher N is better)
        V1s (list) : a length N list containing initial points from our object (higher N is better)
    
    Returns:
        axis (ndarray) : a size (3,) ndarray giving the normalized axis of rotation
        theta (float) : a float giving the angle of rotation theta in radians
    """
    v11, v12, v13 = V1s
    v21, v22, v23 = V2s
    com1 = (v11 + v12 + v13) / 3
    com2 = (v21 + v22 + v23) / 3
    V1 = np.vstack([v - com1 for v in V1s])
    V2 = np.vstack([v - com2 for v in V2s])
    U, S, V = np.linalg.svd(V1.T.dot(V2))
    R = V.dot(U.T)
    if np.linalg.det(R) < 0:
        V[:,2] *= -1
        R = V.dot(U.T)
    theta = -acos((np.trace(R)-1)/2)
    axis = np.array([R[2,1]-R[1,2], R[0,2]-R[2,0], R[1,0]-R[0,1]])
    axis /= np.linalg.norm(axis)
    return axis, theta
    
def get_translation(com1, com2):
    """
    Calculates the translation of the center of mass.
    """
    return com2 - com1
```

### 5.1 Unit testing our simulation
Seems pretty standard. Let's make sure that our model matches other systems if we use the same input parameters. We'll try one of the classic figure-of-eight patterns since they're easy to find parameters for. First, we'll define a helper function:

```python
def solve(q, q_dot, k, T):
    t0 = 0
    z0 = np.hstack(q + q_dot)

    backend = "dop853"
    solver = ode(three_body_2d).set_integrator(backend)

    sol = []
    def solout(t, z):
        sol.append([t, *z])

    solver.set_solout(solout)
    solver.set_initial_value(z0, t0).set_f_params(k)
    solver.integrate(T)

    sol = np.array(sol)

    ts = sol[:,0]
    zs = sol[:,1:]

    X1_p, X2_p, X3_p = zs[:,:2], zs[:,2:4], zs[:,4:6]
    return ts, (X1_p, X2_p, X3_p)
```

Now, let's actually solve the system:

### 5.3 Looking at the inverse problem
The inverse problem takes a bit of thought. To properly scope things, we want to know if -- when we rotate a 2-simplex about it's center of mass and translate it through 3D space -- this corresponds to an orbital trajectory in the projected space.

First, we need to ask ourselves if all rotations of the 3D simplex result in 2D trajectories that could potentially be orbital systems, to which the answer is clearly no. For example, if we rotate our simplex about the $x$ or $y$ axes, we won't get orbitals. Depending on the attitude of the simplex, we'll get either an invalid solution (three straight lines), a two-body collision (axis of rotation orthogonal to the line formed by any two vertices), or a three-body collision (all points collinear in 2D, and we rotate in a direction orthogonal to the line they form). The axis of rotation needs a component that is orthogonal to the projection plane in order to even have a chance of being an orbital trajectory. Secondly, if we translate the simplex along the $z$-axis, we'll get an obvious three-body collision when $z_s \rightarrow \infty$ (if we define positive as down). We'll get a similar blow up if the simplex moves towards us, flinging bodies out to infinity in the 2D space.

These limits are consistent with Sundman's proof that the only collisions occur when the angular momentum $L=0$. If the axis of rotation has no component orthogonal to the $xy$ plane, the rotation of the simplex isn't projected into that plane at any point in time, so the angular momentum vanishes. Similarly, if the simplex moves away from us towards infinity, $L$ vanishes as the vertex points converge to a singularity. Sundman also showed that if $L \neq 0$, there will be no collisions, which makes sense if we view our system as a projected simplex. The points are all physically separated by geometry; by definition it's impossible for them to collide if we're seeing the shadow of a 2-simplex, unless we consider the aforementioned scenarios.

#### 5.3.1 Trying a simple case
Okay, so now that we've thought about that a bit, can we come up with some simple examples that might be stable solutions, by coming up with a periodic rotation and translation of our simplex, and projecting it onto the 2D plane?

First off, let's write a quick function to help us out. This function will take a simplex, a list of periodic displacements to the CG, and a sequence of rotation axes (unit vectors), theta displacements at each timestep, and the number of steps. It'll return the projected positions of the vertices, and the projected center of mass of the simplex at each timestep.

```python
def simulate(simplex, dXs, axes, dthetas, steps):
    K = simplex
    vs, com, norm = get_simplex2(K[0], K[1], K[2])
    Q = [projection(k) for k in K]
    com_p = projection(com)
    
    # array to store position in projected space (i.e. 2d xy coords)
    X1_p = np.zeros((steps, 2))
    X2_p = np.zeros((steps, 2))
    X3_p = np.zeros((steps, 2))
    COM_p = np.zeros((steps, 2))

    # initialize first element of position array to initial position
    X1_p[0,:] = Q[0]
    X2_p[0,:] = Q[1]
    X3_p[0,:] = Q[2]
    COM_p[0,:] = com_p
    
    for t in range(steps-1):
        # move points to origin
        K = [k - com for k in K]

        # rotate and translate our points
        R = np.eye(3)
        for axis, dtheta in zip(axes, dthetas):
            R = R.dot(rotate_around_axis(axis, dtheta))
        K = [R.dot(k.reshape(-1,1)).flatten() for k in K]
        K = [k + com for k in K]
        for dX in dXs:
            K = [k + dX[t] for k in K]
        
        vs, com, norm = get_simplex2(K[0], K[1], K[2])

        # project points and com onto plane
        Q = [projection(k) for k in K]
        com_p = projection(com)

        # save data
        X1_p[t+1,:] = Q[0]
        X2_p[t+1,:] = Q[1]
        X3_p[t+1,:] = Q[2]
        COM_p[t+1,:] = com_p
    
    Xs = [X1_p, X2_p, X3_p]
    return Xs, COM_p

def generate_simplex(random=True):
    v1 = np.array([0, 1/sqrt(3), 0])
    v2 = np.array([-0.5, -1/(2*sqrt(3)), 0])
    v3 = np.array([0.5, -1/(2*sqrt(3)), 0])
    com = get_com(v1, v2, v3)

    # set initial points
    K = [v1, v2, v3]
    dz = np.random.uniform(2, 3)
    if random:
        axis, theta = np.random.uniform(-1, 1, (3,)) ,np.random.uniform(-pi, pi)
        axis /= np.linalg.norm(axis)

        # rotate our points
        R = rotate_around_axis(axis, theta)

        # rotate and translate our points. We'll center it at x=0, y=0
        K = [R.dot(k.reshape(-1,1)).flatten() for k in K]
    com = get_com(K[0], K[1], K[2])
    for k in K:
        k[0] -= com[0]
        k[1] -= com[1]
        k[2] += dz - com[2]
    vs, com, norm = get_simplex2(K[0], K[1], K[2])
    return vs, com, norm
```

Next, we'll create a new simplex, and give it some random attitude in 3D space. Then, we'll rotate it about the z-axis and animate the trajectory. I played around with the timestep a bit, and it seems that we can get away with a much larger timestep in the simplex space, maybe because things are a bit more well-behaved.

```python
K, com, norm = generate_simplex()

print("2-Simplex Parameters: ")
print("Vertex Locations: ")
print("V1: ", K[0])
print("V2: ", K[1])
print("V3: ", K[2])
print()
print("Centre of Mass: ", com)
print("Surface Normal: ", norm)

T = 5
period = 2.5
dt = 1e-1
steps = int(T/dt)
ts = np.linspace(0, T, steps)

# axis of rotation -- we'll use the z-axis first
axis = np.array([0., 0., 1.])
dtheta = 2*pi*dt/period
dX = [np.zeros((3)) for _ in range(steps-1)]

Xs, COM_p = simulate(K, [dX], [axis], [dtheta], steps)
anim = animate_orbit(ts, Xs, COM_p, xlim=(-1.5,1.5), ylim=(-1.5,1.5))
HTML(anim.to_html5_video())
```
